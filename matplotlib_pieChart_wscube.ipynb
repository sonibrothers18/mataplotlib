```python
import matplotlib.pyplot as plt
import numpy as np
```

# Pie Chart -->

```python
x = [10,20,30,40]

plt.pie(x)
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.title("Pie Chart", fontsize = 30)

plt.pie(x, labels=y)
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y, labeldistance = 1.5) # change the distance of label from pie chart
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

ex = [0.0,0.2,0.0,0.0]

plt.pie(x, labels=y, explode=ex)
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

c = ["r", "b", "g", "y"]

plt.pie(x, labels=y, colors=c) # change the color according to your preferences
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y, autopct="%0.2f%%") # autopct used to known the % occupy by each part
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y, shadow=True)
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y, radius=0.5) # change the radius of pie chart
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y, startangle=180) # startangle change the starting angle(point) of parts
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y, textprops = {"fontsize": 25, "color": "green"}) # change the fontsize, color of texts
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y, counterclock = False) # change the direction of parts to clockwise
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y, wedgeprops = {"edgecolor": "black", "linewidth": 5}) # change edge color and it's width
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y, rotatelabels=True) # rotate the lables
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y)
plt.legend(loc=2)

plt.show()
```

```python
plt.pie([1]) # Dot pie chart
plt.show()
```

```python
x = [10,20,30,40]
y = ["c", "c++", "java", "python"]

plt.pie(x, labels=y)
plt.pie([1], radius=0.8, colors="w") # similarly we can compare two piechart together

plt.show()
```
