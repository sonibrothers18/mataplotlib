```python
import matplotlib.pyplot as plt
import numpy as np
```

# Stem Plot -->

```python
x = [1,2,3,3,5,6]
y = [2,2,5,7,3,8]

plt.stem(x, y)
plt.show()
```

```python
x = [1,2,3,3,5,6]
y = [2,2,5,7,3,8]

plt.stem(x, y, orientation="horizontal") # change the graph to horizontally
plt.show()
```

```python
x = [1,2,3,3,5,6]
y = [2,2,5,7,3,8]

plt.stem(x, y, linefmt=":r") # change the style & colour of line
plt.show()
```

```python
x = [1,2,3,3,5,6]
y = [2,2,5,7,3,8]

plt.stem(x, y, markerfmt="r+") # change the style & color of dots
plt.show()
```

```python
x = [1,2,3,3,5,6]
y = [2,2,5,7,3,8]

plt.stem(x, y, bottom=2) # change the value of bottom line to 2
plt.show()
```

```python
x = [1,2,3,3,5,6]
y = [2,2,5,7,3,8]

plt.stem(x, y, basefmt=":g") # change the colour & style of bottom line
plt.show()
```

```python
x = [1,2,3,3,5,6]
y = [2,2,5,7,3,8]

plt.stem(x, y, label="python") # use legend to see this on graph
plt.legend()
plt.show()
```

```python
# REMOVED FROM MATPLOTLIB
x = [1,2,3,3,5,6]
y = [2,2,5,7,3,8]

plt.stem(x, y, use_line_collection=False)
plt.show()
```

```python
 x = [1,2,3,3,5,6]
y = [2,2,5,7,3,8]
z = [5,2,-5,3,3,-8]

plt.stem(x, y, linefmt=":r", label="xy", markerfmt="+")
plt.stem(x, z, linefmt="-.g", label="xz", markerfmt="*")
plt.legend()
plt.show()
```
