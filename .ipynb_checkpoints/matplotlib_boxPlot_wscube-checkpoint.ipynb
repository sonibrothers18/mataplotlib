```python
import matplotlib.pyplot as plt
```

# Box Plot -->

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, vert=False) # change box to horizontal
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, notch=True) # change the apperance of box
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, widths=0.5) # change the width of box
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, tick_labels=["python"]) # change the name of box on x-axis (always pass in list)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, patch_artist=True) # filled the box with colour
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, showmeans=True) # put a triangle on mean of the box
plt.show()
```

## Whisker -->

```python
x = [10,20,30,40,50,60,70,150] # show the hollow circle(whisker)
# It is shown because the difference between 70 and 150 is so much

plt.boxplot(x)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,150]

plt.boxplot(x, whis=3.5) # use to join whisker with line
plt.show()
```

```python
x = [10,20,30,40,50,60,70,150]

plt.boxplot(x, sym="+r") # change the colour and style of whisker(leave empty to hide the whisker)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, boxprops=dict(color="r"), capprops=dict(color="g"), whiskerprops=dict(color="blue"))
# change the colour of box outline, max/min lines, straight line (always give values in dictionary)

plt.show()
```

```python
# DON'T KNOWN THE USE OF THIS

x = [10,20,30,40,50,60,70]

plt.boxplot(x, flierprops=dict(markeredgecolor = "y"))

plt.show()
```

```python
x = [10,20,30,40,50,60,70]
x1 = [10,20,40,50,30,60,90]

y = [x, x1] # if we don't do this, boxes will overlap each other

plt.boxplot(y, tick_labels=["python", "c++"])

plt.show()
```
