```python
import matplotlib.pyplot as plt
import numpy as np
```

# Fill Between Plot -->

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x, y, color="r")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x, y, color="red")

plt.fill_between(x,y, color="lightgreen") # fill graph (colour is optional)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x, y, color="red")

plt.fill_between(x=[2,4], y1=4, y2=2)

plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x, y, color="red")

plt.fill_between(x, y, where=(x>=2) & (x<=4))

plt.show()
```
