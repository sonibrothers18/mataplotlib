```python
import matplotlib.pyplot as plt
import numpy as np
```

# Scatter Plot -->

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]

plt.title("Scatter Plot", fontsize=30)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

plt.scatter(x, y)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]

plt.title("Scatter Plot", fontsize=30)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

colors = ["r", "y", "g", "b", "r", "g", "r"]

plt.scatter(x, y, c=colors) # different colors of dots
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]

plt.title("Scatter Plot", fontsize=30)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

colors = ["r", "y", "g", "b", "r", "g", "r"]

plt.scatter(x, y, c=colors, sizes=[355,100,500]) # different size of dots
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]

plt.title("Scatter Plot", fontsize=30)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

colors = ["r", "y", "g", "b", "r", "g", "r"]

plt.scatter(x, y, c=colors, sizes=[355], alpha=0.5) # alpha is use to change the transparency of dots
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]

plt.title("Scatter Plot", fontsize=30)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

colors = ["r", "y", "g", "b", "r", "g", "r"]

plt.scatter(x, y, c=colors, sizes=[355], marker="*") # marker will change the dots to star
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]

plt.title("Scatter Plot", fontsize=30)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

colors = ["r", "y", "g", "b", "r", "g", "r"]

plt.scatter(x, y, c=colors, sizes=[355], edgecolor="black", linewidth=5) # edgecolor will change the border color (linewidth is optional)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]

plt.title("Scatter Plot", fontsize=30)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

colors = [50,100,5,18,35,80,94]

plt.scatter(x, y, c=colors, sizes=[355], edgecolor="black", cmap="BrBG") #cmap use to choose different color shades options 
plt.colorbar() # use to show color bar

plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]

plt.title("Scatter Plot", fontsize=30)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

colors = [50,100,5,18,35,80,94]

plt.scatter(x, y, c=colors, sizes=[355], edgecolor="black", cmap="BrBG")

temp = plt.colorbar()
temp.set_label("Color Bar") # use to set name of color bar

plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
z = [4,6,9,7,2,3,5]

plt.title("Scatter Plot", fontsize=30)
plt.xlabel("Day", fontsize=15)
plt.ylabel("No.", fontsize=15)

colors1 = [50,100,5,18,35,80,94]
colors2 = ["r"]

plt.scatter(x, y, c=colors1, s=[355], edgecolor="black", cmap="BrBG")
plt.scatter(x, z, c=colors2, s=[355], edgecolor="black")

plt.colorbar()
plt.show()
```
