```python
import matplotlib.pyplot as plt
```

# Axis -->

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x, y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x, y)

plt.xticks(x) # change the values of x-axis (we can also use different values as a list)
plt.yticks(x) # change the values of y-axis

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x, y)

plt.xticks(x, labels=["python", "c", "c++", "java", "rust"])
plt.yticks(x)

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x, y)

plt.xlim(0, 10) # change the limit of graph
plt.ylim(0, 10)

# we can also use this -->
plt.axis([0,10, 0,7])
       # for x, for y

plt.show()
```
