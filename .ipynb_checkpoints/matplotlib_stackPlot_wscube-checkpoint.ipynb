```python
import matplotlib.pyplot as plt
```

# Stack Plot -->

```python
x = [1,2,3,4,5]
area = [1,2,3,4,50]

plt.stackplot(x, area)
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

plt.stackplot(x, area1, area2, area3)
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = ["area1", "area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l) # always use legend
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = ["area1", "area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l, colors=["r", "g", "m"]) # change the colours of graph
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = ["area1", "area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l, baseline="wiggle")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,4]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]

l = ["area1", "area2", "area3"]

plt.stackplot(x, area1, area2, area3, labels=l)

plt.grid() # To show grid line

plt.legend()
plt.show()
```
