```python
import matplotlib.pyplot as plt
import numpy as np
```

# Bar Plot -->

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]

plt.xlabel("Languages", fontsize=20) # fontsize is optional
plt.ylabel("Users", fontsize=20)
plt.title("Popularity", fontsize=50)

plt.bar(x, y)
plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]

plt.xlabel("Languages", fontsize=20)
plt.ylabel("Users", fontsize=20)
plt.title("Popularity", fontsize=50)

plt.barh(x, y) # horizontal bar graph
plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")

plt.bar(x, y, width=0.4) # change the width of bars
plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")

plt.bar(x, y, color="r") # change the color of bar
plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")
c = ["r", "g"]

plt.bar(x, y, color=c) # multiple colors in bar
plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")

plt.bar(x, y, align="edge") # x-axis name will be display on the edge of bar
plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")

plt.bar(x, y, edgecolor = "r", linewidth=5) # edgecolor change the bar boundry to red(linewidth is optional)
plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")

plt.bar(x, y, edgecolor = "r", linewidth=5, linestyle=":") # linestyle change the bar boundry style
plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")

plt.bar(x, y, alpha=0.5) # alpha(0 to 1) is use to dull the color of bars
plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")

plt.bar(x, y, label="lines") # always use lenged with this to display in graph
plt.legend()
plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]
z = [20,30,40,50]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")

plt.bar(x, y, color="r", label="xy")
plt.bar(x, z, color="green", label="xz") # shows 2 different graphs in same
plt.legend()

plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]
z = [20,30,40,50]

# use for making 2 different bar adjacent
width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")

plt.bar(p, y, width, color="r", label="xy")
plt.bar(p1, z, width, color="green", label="xz")
plt.legend()

plt.show()
```

```python
x = ["python", "c", "java", "html"]
y = [80, 45, 48, 12]
z = [20,30,40,50]

# use for making 2 different bar adjacent
width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Languages")
plt.ylabel("Users")
plt.title("Popularity")

plt.bar(p, y, width, color="r", label="xy")
plt.bar(p1, z, width, color="green", label="xz")
plt.legend()

plt.xticks(p+width/2, x, rotation=30) # use to show the names on x-axis & rotation is optional
# width/2 is use to center the name between 2 adjacent bars
plt.show()
```
