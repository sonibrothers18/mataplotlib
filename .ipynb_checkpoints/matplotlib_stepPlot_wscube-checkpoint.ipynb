```python
import matplotlib.pyplot as plt
```

# Step Plot -->

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x, y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x, y)

plt.grid() # To show grid line

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x, y, color="r") # change the colour of line

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x, y, marker="o", ms=20, mfc="r") # add dots on points and changes it's size & colour

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,2,3,5,4]

plt.step(x, y, marker="o", mfc="r", label="Python") # use legend to show label

plt.legend(loc=2) # change the location of label

plt.show()
```
