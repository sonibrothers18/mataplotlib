```python
import matplotlib.pyplot as plt
```

# Savefig -->

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x, y)
plt.savefig("line") # save graph as png with file name "line"
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x, y)
plt.savefig("line1", dpi=2000) # save graph with higher quality as png with file name "line1"
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x, y)
plt.savefig("line2", dpi=1000, facecolor="g") # save graph with green facecolor as png with file name "line2"
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x, y)
plt.savefig("line3.pdf", dpi=1000) # save graph as pdf with file name "line3"
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,3,6,4]

plt.plot(x, y)
plt.savefig("line4", transparent=True) # Transparent the graph and save as "line4"
plt.show()
```
