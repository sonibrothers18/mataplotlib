```python
import matplotlib.pyplot as plt
```

# Text In Plot -->

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x, y)

plt.text(2,5, "java", fontsize=20, style="italic")

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x, y)

plt.text(2,5, "java", bbox={"facecolor" : "red", "alpha":0.5}) # use to create box for text

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x, y)

plt.annotate("python", xy=(2,1), xytext=(3,3), arrowprops=dict(facecolor="red", shrink=100))
    # text to display, arrow coor, text coor, arrow colour

plt.show()
```

```python
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt.plot(x, y)

plt.legend(["up"], facecolor="r", edgecolor="g", framealpha=0.2, loc=4, shadow=True) # loc value is between 0 to 10

plt.show()
```
